## Section 2 — Cleaning and Transformation

```python
# Drop the 3 rows with missing values in artists, album_name, track_name
df = df.dropna(subset=['artists', 'album_name', 'track_name'])

print("Shape after dropping missing rows:", df.shape)
```

```python
# Drop the unnecessary 'Unnamed: 0' column
df = df.drop(columns=['Unnamed: 0'])

print("Columns after drop:")
print(df.columns.tolist())
```

```python
# Convert duration from milliseconds to seconds (new column)
df['duration_sec'] = pd.to_numeric(df['duration_ms'] / 1000, errors='coerce')

print("Duration in seconds (first 5):")
print(df['duration_sec'].head())
```

```python
# Convert explicit column to integer (True=1, False=0)
df['explicit'] = df['explicit'].astype(int)

print("Explicit column after conversion:")
print(df['explicit'].value_counts())
```

```python
# Convert track_genre to categorical dtype
df['track_genre'] = df['track_genre'].astype('category')

print("track_genre dtype:", df['track_genre'].dtype)
```

```python
# Derive new column using np.where
# Categorize songs as 'high' or 'low' popularity
df['popularity_tier'] = np.where(df['popularity'] >= 50, 'high', 'low')

print("Popularity tier counts:")
print(df['popularity_tier'].value_counts())
```

```python
# Derive new column using .apply
# Categorize tempo into slow, medium, fast
def categorize_tempo(tempo):
    if tempo < 90:
        return 'slow'
    elif tempo < 130:
        return 'medium'
    else:
        return 'fast'

df['tempo_category'] = df['tempo'].apply(categorize_tempo)

print("Tempo category counts:")
print(df['tempo_category'].value_counts())
```

```python
# Fill any remaining nulls in numeric columns with median
numeric_cols = df.select_dtypes(include='number').columns
df[numeric_cols] = df[numeric_cols].fillna(df[numeric_cols].median())

print("Missing values after cleaning:")
print(df.isnull().sum().sum())
```

```python
# Final check on cleaned dataframe
print("Final shape:", df.shape)
print("Dtypes:")
print(df.dtypes)
```

```python
# Export cleaned data to data/processed/
df.to_csv("../data/processed/spotify_cleaned.csv", index=False)

print("Cleaned dataset exported successfully!")
```

## Key Findings — Section 2

- Dropped **3 rows** with missing values in artists, album_name, track_name.
- Dropped **Unnamed: 0** column — not needed for analysis.
- Converted **duration_ms** to **duration_sec** for easier interpretation.
- Converted **explicit** from bool to int (1 = explicit, 0 = not explicit).
- Converted **track_genre** to categorical dtype for memory efficiency.
- Created **popularity_tier** column using np.where — high (>=50) vs low (<50).
- Created **tempo_category** column using .apply — slow/medium/fast.
- Filled remaining numeric nulls with median values.
- Cleaned dataset exported to **data/processed/spotify_cleaned.csv**.

### Research Q4 - Is There a Song Length “Sweet Spot” for Popularity?

```python
# Reload cleaned dataset for Section 4
df = pd.read_csv("../data/processed/spotify_cleaned.csv")
print(df.shape)

# Create quantile-based duration buckets
duration_labels = ['Q1 shortest', 'Q2', 'Q3', 'Q4', 'Q5 longest']
df['duration_quantile'] = pd.qcut(df['duration_sec'], q=5, labels=duration_labels)

duration_stats = (
    df.groupby('duration_quantile', observed=False)['popularity']
      .agg(['mean', 'median', 'count'])
      .round(3)
      .reset_index()
 )

print("Average popularity by song length quantile:")
print(duration_stats)

best_bucket = duration_stats.loc[duration_stats['mean'].idxmax(), 'duration_quantile']
print()
print(f"Most popular duration bucket: {best_bucket}")
```

```python
# Bar plot — average popularity by duration quantile
fig, ax = plt.subplots(figsize=(10, 5))

bars = ax.bar(
    duration_stats['duration_quantile'],
    duration_stats['mean'],
    color=['#6baed6', '#9ecae1', '#c6dbef', '#fdae6b', '#fd8d3c'],
    edgecolor='white'
 )

ax.set_title('Average Popularity by Song Length Quantile', fontsize=14)
ax.set_xlabel('Duration Quantile')
ax.set_ylabel('Average Popularity')
ax.bar_label(bars, fmt='%.2f', padding=3)
ax.grid(axis='y', alpha=0.3)

fig.tight_layout()
fig.savefig("../figures/popularity_by_duration_quantile.png", dpi=300)
plt.show()
print("Figure saved: popularity_by_duration_quantile.png")
```

### Research Q5 - Do Collaborations (Multiple Artists) Score Higher Than Solo Tracks?

```python
# Create a simple collaboration feature from the artists column
artist_split_pattern = r';|,|&| feat\.| featuring '
df['artist_count'] = (
    df['artists']
      .fillna('')
      .str.split(artist_split_pattern, regex=True)
      .apply(lambda x: len([name.strip() for name in x if name.strip()]))
 )

# Safety fallback for malformed strings
df['artist_count'] = df['artist_count'].replace(0, 1)
df['collab_type'] = np.where(df['artist_count'] > 1, 'Collaboration', 'Solo')

collab_stats = (
    df.groupby('collab_type')['popularity']
      .agg(['mean', 'median', 'count'])
      .round(3)
 )

print("Popularity by track type (solo vs collaboration):")
print(collab_stats)

artist_count_stats = (
    df[df['artist_count'] <= 6]
      .groupby('artist_count')['popularity']
      .agg(['mean', 'count'])
      .round(3)
      .reset_index()
 )

print("\nPopularity by number of credited artists (1 to 6):")
print(artist_count_stats)
```

```python
# Visualization 1 — solo vs collaboration average popularity
fig, axes = plt.subplots(1, 2, figsize=(14, 5))

bars = axes[0].bar(
    collab_stats.index,
    collab_stats['mean'],
    color=['#4C78A8', '#F58518'],
    edgecolor='white'
 )
axes[0].set_title('Average Popularity: Solo vs Collaboration', fontsize=13)
axes[0].set_xlabel('Track Type')
axes[0].set_ylabel('Average Popularity')
axes[0].bar_label(bars, fmt='%.2f', padding=3)
axes[0].grid(axis='y', alpha=0.3)

# Visualization 2 — popularity by artist count (1–6)
axes[1].plot(
    artist_count_stats['artist_count'],
    artist_count_stats['mean'],
    marker='o',
    linewidth=2,
    color='#1DB954'
 )
axes[1].set_title('Popularity by Number of Credited Artists (1–6)', fontsize=13)
axes[1].set_xlabel('Number of Artists')
axes[1].set_ylabel('Average Popularity')
axes[1].set_xticks([1, 2, 3, 4, 5, 6])
axes[1].grid(alpha=0.3)

fig.tight_layout()
fig.savefig("../figures/popularity_by_collaboration.png", dpi=300)
plt.show()
print("Figure saved: popularity_by_collaboration.png")
```
