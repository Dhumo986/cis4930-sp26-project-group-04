## Section 2 — Cleaning and Transformation

```python
# Drop the 3 rows with missing values in artists, album_name, track_name
df = df.dropna(subset=['artists', 'album_name', 'track_name'])

print("Shape after dropping missing rows:", df.shape)
```

```python
# Drop the unnecessary 'Unnamed: 0' column
df = df.drop(columns=['Unnamed: 0'])

print("Columns after drop:")
print(df.columns.tolist())
```

```python
# Convert duration from milliseconds to seconds (new column)
df['duration_sec'] = pd.to_numeric(df['duration_ms'] / 1000, errors='coerce')

print("Duration in seconds (first 5):")
print(df['duration_sec'].head())
```

```python
# Convert explicit column to integer (True=1, False=0)
df['explicit'] = df['explicit'].astype(int)

print("Explicit column after conversion:")
print(df['explicit'].value_counts())
```

```python
# Convert track_genre to categorical dtype
df['track_genre'] = df['track_genre'].astype('category')

print("track_genre dtype:", df['track_genre'].dtype)
```

```python
# Derive new column using np.where
# Categorize songs as 'high' or 'low' popularity
df['popularity_tier'] = np.where(df['popularity'] >= 50, 'high', 'low')

print("Popularity tier counts:")
print(df['popularity_tier'].value_counts())
```

```python
# Derive new column using .apply
# Categorize tempo into slow, medium, fast
def categorize_tempo(tempo):
    if tempo < 90:
        return 'slow'
    elif tempo < 130:
        return 'medium'
    else:
        return 'fast'

df['tempo_category'] = df['tempo'].apply(categorize_tempo)

print("Tempo category counts:")
print(df['tempo_category'].value_counts())
```

```python
# Fill any remaining nulls in numeric columns with median
numeric_cols = df.select_dtypes(include='number').columns
df[numeric_cols] = df[numeric_cols].fillna(df[numeric_cols].median())

print("Missing values after cleaning:")
print(df.isnull().sum().sum())
```

```python
# Final check on cleaned dataframe
print("Final shape:", df.shape)
print("Dtypes:")
print(df.dtypes)
```

```python
# Export cleaned data to data/processed/
df.to_csv("../data/processed/spotify_cleaned.csv", index=False)

print("Cleaned dataset exported successfully!")
```

## Key Findings — Section 2

- Dropped **3 rows** with missing values in artists, album_name, track_name.
- Dropped **Unnamed: 0** column — not needed for analysis.
- Converted **duration_ms** to **duration_sec** for easier interpretation.
- Converted **explicit** from bool to int (1 = explicit, 0 = not explicit).
- Converted **track_genre** to categorical dtype for memory efficiency.
- Created **popularity_tier** column using np.where — high (>=50) vs low (<50).
- Created **tempo_category** column using .apply — slow/medium/fast.
- Filled remaining numeric nulls with median values.
- Cleaned dataset exported to **data/processed/spotify_cleaned.csv**.

## Section 3 — Descriptive Statistics and EDA

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns

# Reload cleaned dataset
df = pd.read_csv("../data/processed/spotify_cleaned.csv")
print("Loaded successfully!")
print(df.shape)
```

```python
# Reload cleaned dataset
df = pd.read_csv("../data/processed/spotify_cleaned.csv")
print("Loaded successfully!")
print(df.shape)
```

```python
# Descriptive stats on key numeric columns
numeric_cols = ['popularity', 'danceability', 'energy', 
                'loudness', 'tempo', 'valence', 'acousticness',
                'instrumentalness', 'speechiness']

print(df[numeric_cols].describe().round(3))
```

```python
# Explicit stats on individual columns
for col in ['popularity', 'danceability', 'energy']:
    print(f"=== {col} ===")
    print(f"Mean:   {df[col].mean():.3f}")
    print(f"Median: {df[col].median():.3f}")
    print(f"Min:    {df[col].min():.3f}")
    print(f"Max:    {df[col].max():.3f}")
    print(f"Std:    {df[col].std():.3f}")
    print()
```

```python
# Grouped stats by genre — mean popularity
genre_popularity = df.groupby('track_genre')['popularity'].agg(['mean', 'count', 'std'])
genre_popularity = genre_popularity.sort_values('mean', ascending=False)

print("Top 10 genres by average popularity:")
print(genre_popularity.head(10).round(3))
```

```python
# Grouped stats by genre — multiple features
genre_features = df.groupby('track_genre')[['danceability', 'energy', 
                                             'valence', 'popularity']].agg(['mean', 'count'])
print("Genre feature averages (first 5 genres):")
print(genre_features.head().round(3))
```

```python
# Grouped stats by explicit flag
explicit_stats = df.groupby('explicit')[['popularity', 'danceability', 
                                          'energy', 'valence']].agg(['mean', 'count'])
print("Stats by explicit flag (0=clean, 1=explicit):")
print(explicit_stats.round(3))
```

```python
# Grouped stats by tempo category
tempo_stats = df.groupby('tempo_category')[['popularity', 'danceability', 
                                             'energy']].agg(['mean', 'count'])
print("Stats by tempo category:")
print(tempo_stats.round(3))
```

```python
# groupby with multiple aggregations — required by rubric
grouped = df.groupby(['track_genre', 'tempo_category'])['popularity'].agg(['mean', 
                                                                             'count',
                                                                             'std'])
grouped = grouped.sort_values('mean', ascending=False)
print("Top 10 genre + tempo combinations by popularity:")
print(grouped.head(10).round(3))
```

```python
# Pivot table — average popularity by genre and tempo category
pivot = df.groupby(['track_genre', 'tempo_category'])['popularity'].mean()
pivot = pivot.unstack(fill_value=0).round(2)

print("Pivot table — average popularity by genre and tempo:")
print(pivot.head(10))
```

```python
# Correlation matrix
corr_cols = ['popularity', 'danceability', 'energy', 'loudness', 
             'tempo', 'valence', 'acousticness', 
             'instrumentalness', 'speechiness']

corr_matrix = df[corr_cols].corr()
print("Correlation matrix:")
print(corr_matrix.round(3))
```

```python
# Correlation heatmap
fig, ax = plt.subplots(figsize=(10, 8))

sns.heatmap(corr_matrix, 
            annot=True, 
            fmt='.2f',
            cmap='coolwarm',
            center=0,
            ax=ax)

ax.set_title('Correlation Matrix of Audio Features', fontsize=14)
fig.tight_layout()
fig.savefig("../figures/correlation_heatmap.png", dpi=300)
plt.show()
print("Heatmap saved!")
```

```python
# Research Q1 — What features correlate most with popularity?
popularity_corr = corr_matrix['popularity'].sort_values(ascending=False)
print("Features most correlated with popularity:")
print(popularity_corr.round(3))
```

```python
# Research Q2 — Which genre is most energetic vs most relaxed?
genre_energy = df.groupby('track_genre')['energy'].mean().sort_values(ascending=False)

print("Top 5 most energetic genres:")
print(genre_energy.head(5).round(3))
print()
print("Top 5 most relaxed genres:")
print(genre_energy.tail(5).round(3))
```

```python
# Research Q3 — Do explicit songs score higher in popularity?
explicit_pop = df.groupby('explicit')['popularity'].mean()
print("Average popularity by explicit flag:")
print(explicit_pop.round(3))
print()
diff = explicit_pop[1] - explicit_pop[0]
print(f"Explicit songs score {diff:.2f} points higher on average")
```

## Key Findings — Section 3

### Descriptive Statistics
- Average popularity is **33.26** out of 100 — wide spread across tracks
- Average danceability is **0.57** — most songs are moderately danceable
- Average energy is **0.64** — dataset leans towards high energy songs
- Loudness ranges from **-49.53 dB** to **1.80 dB**
- Average tempo is **122 BPM** — most songs fall in medium tempo range

### Grouped Stats by Genre
- Most popular genre on average: **pop-film (58.45)** followed by **k-pop (56.70)**
- Most energetic genres: **death-metal, grindcore, metalcore, happy, hardstyle**
- Most relaxed genres: **classical, new-age, ambient, disney, romance**
- Pop-film and k-pop dominate popularity regardless of tempo category

### Grouped Stats by Explicit Flag
- Explicit songs average popularity: **36.61**
- Clean songs average popularity: **32.95**
- Explicit songs score **3.66 points higher** on average
- Explicit songs also have higher danceability (0.636 vs 0.559) and energy (0.720 vs 0.634)

### Grouped Stats by Tempo
- Fast tempo songs are slightly more popular (**33.74**) than medium (**33.11**) and slow (**32.52**)
- Medium tempo dominates with **24,481 songs** — most songs fall in 90-130 BPM range
- Slow songs have noticeably lower danceability (**0.478**) than fast (**0.529**)

### Correlation Analysis
- Strongest positive correlation with popularity: **loudness (0.053)**
- Strongest negative correlation with popularity: **instrumentalness (-0.093)**
- **Energy and loudness** are strongly correlated (0.761) — louder songs are more energetic
- **Acousticness and energy** are strongly negatively correlated (-0.735) — acoustic songs tend to be less energetic
- Overall popularity correlations are weak — no single feature strongly predicts popularity

### Key Insight
Popularity on Spotify is hard to predict from audio features alone — the correlations 
are all weak (below 0.10). However, explicit songs, louder songs, and tracks from 
pop-film and k-pop genres tend to perform better. The strongest relationships in the 
data are between technical audio features like energy, loudness, and acousticness — 
not between those features and popularity itself.

## Section 4 — Visualizations and Feature Exploration

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns

df = pd.read_csv('../data/processed/spotify_cleaned.csv')
print(df.shape)
```

```python
# line plot — avg popularity by genre (top 20, sorted)
# no release year column in dataset, so genre is used as the ordered x-axis
genre_pop = df.groupby('track_genre')['popularity'].mean().sort_values(ascending=False).head(20)

fig, ax = plt.subplots(figsize=(14, 6))
ax.plot(range(len(genre_pop)), genre_pop.values, marker='o', linewidth=2, color='tab:blue')
ax.set_xticks(range(len(genre_pop)))
ax.set_xticklabels(genre_pop.index, rotation=45, ha='right')
ax.set_xlabel('Genre')
ax.set_ylabel('Average Popularity')
ax.set_title('Average Popularity by Genre (Top 20)')
ax.grid(axis='y', alpha=0.3)
fig.tight_layout()
fig.savefig('../figures/popularity_by_genre.png', dpi=300)
plt.show()
```

```python
# bar plot — avg energy by genre (top 15)
genre_energy = df.groupby('track_genre')['energy'].mean().sort_values(ascending=False).head(15)

fig, ax = plt.subplots(figsize=(12, 6))
ax.bar(range(len(genre_energy)), genre_energy.values,
       color=sns.color_palette('viridis', len(genre_energy)))
ax.set_xticks(range(len(genre_energy)))
ax.set_xticklabels(genre_energy.index, rotation=45, ha='right')
ax.set_xlabel('Genre')
ax.set_ylabel('Average Energy')
ax.set_title('Average Energy by Genre (Top 15)')
ax.set_ylim(0, 1)
ax.grid(axis='y', alpha=0.3)
fig.tight_layout()
fig.savefig('../figures/energy_by_genre.png', dpi=300)
plt.show()
```

```python
# histogram — danceability distribution
fig, ax = plt.subplots(figsize=(10, 6))
ax.hist(df['danceability'], bins=40, color='steelblue', edgecolor='white', alpha=0.8)
ax.axvline(df['danceability'].mean(), color='red', linestyle='--', linewidth=2,
           label=f"Mean: {df['danceability'].mean():.3f}")
ax.axvline(df['danceability'].median(), color='orange', linestyle='--', linewidth=2,
           label=f"Median: {df['danceability'].median():.3f}")
ax.set_xlabel('Danceability')
ax.set_ylabel('Frequency')
ax.set_title('Distribution of Danceability')
ax.legend()
fig.tight_layout()
fig.savefig('../figures/danceability_distribution.png', dpi=300)
plt.show()
```

```python
# scatter plot — energy vs loudness
fig, ax = plt.subplots(figsize=(10, 7))
ax.scatter(df['energy'], df['loudness'], alpha=0.1, s=5, edgecolors='none', color='tab:blue')
ax.set_xlabel('Energy')
ax.set_ylabel('Loudness (dB)')
ax.set_title('Energy vs. Loudness')
fig.tight_layout()
fig.savefig('../figures/energy_vs_loudness.png', dpi=300)
plt.show()
```
